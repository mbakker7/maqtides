```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from tidal_maq import tidalmaq
```

Example with two aquifers and two leaky layers

```python
T = np.array([100, 200])  # m^2/d
c = np.array([500, 2000]) # d
Sll = np.array([1e-4, 2e-4]) # -
Saq = np.array([4e-4, 8e-4]) # -
beta = 0.8 * np.ones(2)
gamma = 0.8 * np.ones(2)
tau = 0.5 # d

#MAQ
mlmaq = tidalmaq(Tsea=T, Tland=T, Ssea=Saq, Sland=Saq, 
                 csea=c, cland=c, sigsea=Sll, sigland=Sll, 
                 beta=beta, gamma=gamma, tau=tau, hs=1)
```

```python
x = np.linspace(-500, 500, 100)
phase = mlmaq.phase(x)
plt.plot(x, phase[0], label='layer 0')
plt.plot(x, phase[1], label='layer 1')
plt.title('phase')
plt.xlabel('x (m)')
plt.ylabel('phase shift (d)')
plt.legend();
```

```python
x = np.linspace(-500, 500, 100)
phase = mlmaq.amp(x)
plt.plot(x, phase[0], label='layer 0')
plt.plot(x, phase[1], label='layer 1')
plt.title('amplitude')
plt.xlabel('x (m)')
plt.ylabel('amplitude (m)')
plt.legend();
```
